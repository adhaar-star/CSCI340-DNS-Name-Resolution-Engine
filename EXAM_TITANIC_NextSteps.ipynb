#MODEL BUILDING

##Try a simple model first - e.g., Logistic Regression

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn import utils

titanic.Sex = pd.Categorical(titanic.Sex)
titanic['Sex'] = titanic.Sex.cat.codes

titanic.Embarked = pd.Categorical(titanic.Embarked)
titanic['Embarked'] = titanic.Embarked.cat.codes

y_data1 = titanic['Survived'].values
X_label =  titanic.drop(columns=['Cabin','Name','Ticket','Survived','Embarked']);
print(X_label.columns)
print(X_label)

x_data1 = titanic[X_label.columns].values

print(x_data1)
print(y_data1)
x_training_data, x_test_data, y_training_data, y_test_data = train_test_split(x_data1, y_data1, test_size=0.33, random_state=0)
model = LogisticRegression(C=300.0, random_state=0)
model.fit(x_training_data, y_training_data)
```

###Train, Test, Evaluate

```
print("Performance on training data: {:.2f}".format(model.score(x_training_data,y_training_data)))
from sklearn.metrics import classification_report
predictions = model.predict(x_test_data)
print(classification_report(y_test_data, predictions))
```

###Try to Improve on this model

```
improved_model = LogisticRegression(C=1.0, random_state=0)
improved_model.fit(x_training_data, y_training_data)
print("Performance on training data: {:.2f}".format(improved_model.score(x_training_data,y_training_data)))
```

```
from sklearn import svm

classifier=svm.SVC(kernel='linear')
classifier.fit(x_training_data, y_training_data.ravel())

from sklearn.metrics import classification_report
svmpredivt = classifier.predict(x_test_data)
print(classification_report(y_test_data, svmpredivt))
```

```
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(random_state=40)


tree.fit(x_training_data, y_training_data)
print("Accuracy on training set: {:.3f}".format(tree.score(x_training_data, y_training_data)))
print("Accuracy on test set: {:.3f}".format(tree.score(x_test_data, y_test_data)))
```

```
tuned_parameters = [{'solver':['lbfgs','adam','sgd'],'activation': ['tanh','relu'], 'hidden_layer_sizes': [6], 'max_iter':[300]},
{'solver':['lbfgs','adam','sgd'],'activation': ['tanh','relu'], 'hidden_layer_sizes': [6,6,6], 'max_iter':[300]}]
```

```
for score in scores:
  print("# Tuning hyper-parameters for %s" % score)
  print()
  clf = GridSearchCV(MLPClassifier(), tuned_parameters, scoring='%s_macro' % score)
  clf.fit(x_training_data,y_training_data)

  print("Best parameters set found on development set:")
  print()
  print(clf.best_params_)
  print()
  print("Grid scores on development set:")
  print()
  means = clf.cv_results_['mean_test_score']
  stds = clf.cv_results_['std_test_score']
  for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r"
        % (mean, std*2, params))
    print()
    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test_data, clf.predict(x_test_data)
    print(classification_report(y_true, y_pred))
    print()
```
